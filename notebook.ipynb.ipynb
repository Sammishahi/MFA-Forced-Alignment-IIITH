```python
#lower case as  txt files 
import os

txt_path = "/kaggle/working/Assignment/transcripts"

for f in os.listdir(txt_path):
    old = os.path.join(txt_path, f)
    new = os.path.join(txt_path, f.lower())
    
    if old != new:
        os.rename(old, new)

print("TXT files normalized.")
```

**prepare MFA FORMAT**

```python
import shutil, os

wav = "/kaggle/working/Assignment/wav"
txt = "/kaggle/working/Assignment/transcripts"

out = "/kaggle/working/mfa_data/speaker1"
os.makedirs(out, exist_ok=True)

for f in os.listdir(wav):
    name = f.replace(".wav","")
    
    shutil.copy(wav+"/"+f, out+"/"+f)
    shutil.copy(txt+"/"+name+".txt", out+"/"+name+".txt")
```
